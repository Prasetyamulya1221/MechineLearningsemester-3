```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# --- contoh kalau punya file CSV ---
# df = pd.read_csv("namafile.csv")

# --- contoh kalau mau data manual ---
df = pd.DataFrame({
    "Height": [150, 155, 160, 162, 165, 168, 170, 172, 175, 180, 182, 185]
})

# Ambil data Height
data_height = df["Height"]

# Buat histogram
n, bins, patches = plt.hist(data_height, bins=5, color='pink', edgecolor='black')

# Tambahkan label
plt.title('Histogram Nilai')
plt.xlabel('Height')
plt.ylabel('Frekuensi')

# Tampilkan rentang frekuensi di sumbu x
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.xticks(bin_centers, ['{:.0f}-{:.0f}'.format(bins[i], bins[i+1]) for i in range(len(bins)-1)])

# Tampilkan histogram
plt.show()
```
