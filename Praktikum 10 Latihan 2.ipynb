```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "NIM": ["TI001","TI002","TI003","TI004","TI005",
            "TI006","TI007","TI008","TI009","TI010"],

    "Actual": [
        "Lulus", "Lulus", "Lulus", "Lulus", "Lulus",
        "Tidak Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus"
    ],

    "Predicted": [
        "Lulus", "Lulus", "Lulus", "Tidak Lulus", "Tidak Lulus",
        "Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus"
    ]
}

df = pd.DataFrame(data)
print("=== Dataset Latihan 2 ===")
print(df)
```

```python
df["Actual_enc"] = df["Actual"].map({"Lulus": 1, "Tidak Lulus": 0})
df["Predicted_enc"] = df["Predicted"].map({"Lulus": 1, "Tidak Lulus": 0})

y_true = df["Actual_enc"]
y_pred = df["Predicted_enc"]
```

```python
cm = confusion_matrix(y_true, y_pred)
print("\n=== Confusion Matrix ===")
print(cm)
```

```python
# Visualisasi Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Pred: Tidak Lulus", "Pred: Lulus"],
            yticklabels=["Actual: Tidak Lulus", "Actual: Lulus"])
plt.title("Confusion Matrix â€“ Klasifikasi Kelulusan")
plt.show()
```

```python
# 4. Hitung Accuracy, Precision, Recall
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)

print("\n=== Evaluasi Model ===")
print(f"Accuracy : {accuracy * 100:.2f}%")
print(f"Precision: {precision * 100:.2f}%")
print(f"Recall   : {recall * 100:.2f}%")
```
